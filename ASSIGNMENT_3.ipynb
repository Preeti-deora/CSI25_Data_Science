**Dataset Used:**


**Titanic: Machine Learning from Disaster**
Available on Kaggle: https://www.kaggle.com/competitions/titanic/data

We'll visualize various aspects such as survival rate, passenger class distribution, age distribution, and more.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Titanic dataset from GitHub raw link
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
print(df.head())
print(df.info())
```

```python
# Visualization 1: Survival Count
sns.countplot(data=df, x='Survived')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Passenger Count')
plt.show()
```

```python
# Visualization 2: Survival by Gender
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival Count by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 3: Passenger Class Distribution

pclass_counts = df['Pclass'].value_counts()
plt.pie(pclass_counts, labels=['Class 1', 'Class 2', 'Class 3'], autopct='%1.1f%%', startangle=90)
plt.title('Passenger Class Distribution')
plt.axis('equal')
plt.show()
```

```python
# Visualization 4: Age Distribution
plt.hist(df['Age'].dropna(), bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Visualization 5: Age vs Survival
sns.boxplot(data=df, x='Survived', y='Age')
plt.title('Age Distribution by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
# Visualization 6: Missing Data Heatmap
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

```python
# Visualization 7: Correlation Heatmap
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Summary of Insights:**

More passengers died than survived.

Women had a significantly higher survival rate than men.

First class passengers had higher chances of survival.

There is a wide range of ages, but survival tends to favor younger and wealthier passengers.
